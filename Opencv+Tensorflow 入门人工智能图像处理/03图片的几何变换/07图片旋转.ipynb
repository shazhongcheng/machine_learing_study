```python
import cv2
import numpy as np
img=cv2.imread('image0.jpg',1)
imgInfo=img.shape
height=imgInfo[0]
width=imgInfo[1]
deep=imgInfo[2]

# 2*3
matRotate = cv2.getRotationMatrix2D((height*0.5,width*0.5),45,0.5) # mat rotate 1 center 2 angle 3 缩放
#100*100 25
dst=cv2.warpAffine(img,matRotate,(width,height))
cv2.imshow('dst',dst)
cv2.waitKey(0)
```
