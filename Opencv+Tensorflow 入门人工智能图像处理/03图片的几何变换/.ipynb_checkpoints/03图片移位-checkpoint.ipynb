# API warpAffine实现

```python
# 1 API 2 算法原理  3 源代码
import cv2
import numpy as np
img=cv2.imread('image0.jpg',1)
cv2.imshow('src',img)
imgInfo=img.shape
height=imgInfo[0]
width=imgInfo[1]
mode=imgInfo[2]
###
matShift=np.float32([[1,0,100],[0,1,200]])
dst=cv2.warpAffine(img,matShift,(height,width)) # 1 data 2 mat 3 info
# 移位 矩阵
cv2.imshow('dst',dst)
cv2.waitKey(0)
```

# 源代码实现

```python
# [1,0,100],[0,1,200] 2*2 2*1
# [[1,0],[0,1]] 2*2 A
# [[100],[200]] 2*1 B
# x y C
# A*C+B =[1*x+0*y],[0*x+1*y]+[[100],[200]]
# =[[x+100],[y+200]]

# (10,20) -> (110,220)

import cv2
import numpy as np
img=cv2.imread('image0.jpg',1)
cv2.imshow('src',img)
imgInfo=img.shape
height=imgInfo[0]
width=imgInfo[1]
mode=imgInfo[2]
dst = np.zeros(img.shape,np.uint8)
for i in range(0,height):
    for j in range(0,width-100):
        dst[i,j+100]=img[i,j]
cv2.imshow('dst',dst)
cv2.waitKey(0)
```
