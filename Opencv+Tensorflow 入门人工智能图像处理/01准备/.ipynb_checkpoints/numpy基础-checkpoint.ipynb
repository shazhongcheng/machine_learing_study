# CRUD

```python
import numpy as np
data1 = np.array([1,2,3,4])
data2 = np.array([[1,2],[3,4]])
print(data1)
print(data2)

#维度
print(data1.shape,data2.shape)
#zeros ones
print(np.zeros([2,3]),np.ones([2,3]))
#改查
data2[1,0]=5
print(data2)
print(data2[1,1])
#基本运算  对应元素
data3=np.ones([2,3])
print(data3*2) #对应相乘
print(data3/2) #对应
print(data3+2) #对应
print(data3-2) #对应
#矩阵+*  对应元素
data4=np.array([[1,2,3],[4,5,6]])
print(data3+data4)
print(data3*data4)
```
